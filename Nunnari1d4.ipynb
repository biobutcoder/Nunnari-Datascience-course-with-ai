```python
# Accuracy metrics
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score

def error_metrics(y, y_hat):
  print("MSE", mean_squared_error(y, y_hat))
  print("MAE", mean_absolute_error(y, y_hat))
  print("Score", r2_score(y, y_hat))
```

```python
y_hat = model.predict(X)
error_metrics(Y, y_hat)
```

```python
import numpy as np
```

```python
def r2_score(y, y_hat):
  return 1 - np.var(y - y_hat) / np.var(y)
```

```python
r2_score(Y, y_hat)
```

Polynomial Regression

```python
def PlotPolly(model, independent_variable, dependent_variable, Name):
    x_new = np.linspace(independent_variable.min(), independent_variable.max(), 100)
    y_new = model(x_new)

    plt.figure(figsize=(10, 6))
    plt.plot(independent_variable, dependent_variable, '.', label='Data')
    plt.plot(x_new, y_new, '-', label='Fitted Line')  # ← this draws the line
    plt.title(f'Polynomial Fit: {Name}')
    plt.xlabel(Name)
    plt.ylabel('Target Variable')
    plt.legend()
    plt.show()
```

```python
x = df['Living Area']
y = df['price']
```

```python
import numpy as np
x_new = np.linspace(15, 55, 100)
x_new
```

```python
f = np.polyfit(x, y, 1)
p = np.poly1d(f)
print(p)
PlotPolly(p, x, y, 'Living Area')
```

```python
f = np.polyfit(x, y, 3)
p = np.poly1d(f)
print(p)
PlotPolly(p, x, y, 'Living Area')
```

```python
f = np.polyfit(x, y, 10)
p = np.poly1d(f)
print(p)
PlotPolly(p, x, y, 'Living Area')
```
